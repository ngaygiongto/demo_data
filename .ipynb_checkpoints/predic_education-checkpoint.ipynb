```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data_f = pd.read_excel('HSA_HD_THPT_Final.xlsx')
```

```python
data_f.head()
```

## 1. Tổng quan dữ liệu

```python
data_f.info()
```

```python
data_f.isna().sum()
```

```python
data_f.describe()
```

```python
# df = data_f.iloc[:,:-1]
fields = ['Ngày thi', 'Số CMND']
df = data_f.drop(fields, axis = 1)
```

```python
df
```

```python
for col in df.columns:
    missing_data = df[col].isna().sum()
    missing_percent = missing_data/len(df)*100
    print(f'Columns: {col} có {missing_percent}% data missing')
```

```python
df['HAS'].max()
```

## 2. Mối tương quan dữ liệu

```python
sns.set_theme()
```

```python
df['Tên bài thi tổ hợp'].value_counts()
```

```python
df['Giới tính'].value_counts()
```

```python
sns.countplot(data =df, x='Giới tính');
```

```python
df_corr = df.corr()
df_corr
```

```python
sns.heatmap(data=df_corr)
```

```python
df_fg = sns.FacetGrid(data = df, row='Tên bài thi tổ hợp', col='Giới tính')
df_fg.map(sns.scatterplot, 'HAS', 'Ngoại ngữ');
```

```python
new_fg = sns.FacetGrid(data=df, col='Giới tính', palette='Set2',
                       hue='Tên bài thi tổ hợp')
new_fg.map(sns.scatterplot, 'HAS', 'Toán')
```

```python
sns.displot(df, x='HAS', col='Giới tính', kde=True);
```

```python
sns.displot(df, x='HAS', col='Tên bài thi tổ hợp', kde = True)
```
